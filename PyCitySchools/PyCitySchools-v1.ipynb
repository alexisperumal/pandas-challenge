```python
# Alexis Perumal, 11/24/19, PyCitySchools.py
# UCSD Data Science Bootcamp, HW4 - Pandas, Ex. 2
#
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
num_schools = school_data["school_name"].nunique()
num_students = student_data["Student ID"].nunique() # Total number of students from the students dataset, not the
                                                    # reduced number of unique students in the left joined DF.
total_schools_budget = school_data["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()
overall_passing_rate = (avg_math_score+avg_reading_score)/2
percent_passing_math = student_data.loc[(student_data["math_score"]>=70)]["Student ID"].count()*100/num_students
percent_passing_reading = student_data.loc[(student_data["reading_score"]>=70)]["Student ID"].count()*100/num_students

# Format the results
district_summary_df = pd.DataFrame({"Total Schools":[num_schools],
                                   "Total Students":[str(f"{num_students:,}")],
                                    "Total Budget":[str(f"${total_schools_budget:,.2f}")],
                                    "Average Math Score":[avg_math_score],
                                    "Average Reading Score":[avg_reading_score],
                                    "% Passing Math":[percent_passing_math],
                                    "% Passing Reading":[percent_passing_reading],
                                    "% Overall Passing Rate":[overall_passing_rate]
                                   })

# Order the columns correctly
district_summary_df = district_summary_df[["Total Schools", "Total Students", "Total Budget", "Average Math Score",
                                          "Average Reading Score", "% Passing Math", "% Passing Reading",
                                          "% Overall Passing Rate"]]
district_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_summary_gb = school_data_complete.groupby(['school_name'])
school_type = school_summary_gb["type"].unique().str[0]
total_students = school_summary_gb["Student ID"].nunique()
total_school_budget = school_summary_gb["budget"].mean()
per_student_budget = school_summary_gb["budget"].mean() / total_students
avg_math_score = school_summary_gb["math_score"].mean()
avg_reading_score = school_summary_gb["reading_score"].mean()
percent_passing_math = school_data_complete.loc[(school_data_complete["math_score"]>=70)].groupby(['school_name'])\
            ["Student ID"].count() / total_students * 100
percent_passing_reading = school_data_complete.loc[(school_data_complete["reading_score"]>=70)].groupby(['school_name'])\
            ["Student ID"].count() / total_students * 100
overall_passing_rate = (percent_passing_math+percent_passing_reading)/2
# overall_passing_rate
# school_type
```

```python
# Format the results
school_performance_df = pd.DataFrame({"School Type":school_type,
                                     "Total Students":total_students,
                                     "Total School Budget":total_school_budget.map('${:,.2f}'.format),
                                     "Per Student Budget":per_student_budget.map('${:,.2f}'.format),
                                     "Average Math Score":avg_math_score,
                                     "Average Reading Score":avg_reading_score,
                                     "% Passing Math":percent_passing_math,
                                     "% Passing Reading":percent_passing_reading,
                                     "% Overall Passing Rate":overall_passing_rate})
school_performance_df = school_performance_df[["School Type","Total Students", "Total School Budget",
                                               "Per Student Budget", "Average Math Score", "Average Reading Score",
                                               "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
school_performance_df = school_performance_df.sort_values("% Overall Passing Rate",ascending=False)
del school_performance_df.index.name
school_performance_df.head()
```

```python
# Write out the school performance resuts to a csv.
school_performance_df.to_csv("school_performance.csv", index=False, header=True)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a pandas series for each grade. Hint: use a conditional statement. Group each series by school
G9_df = student_data.loc[(student_data["grade"]=="9th")].groupby(["school_name"]).mean()
G10_df = student_data.loc[(student_data["grade"]=="10th")].groupby(["school_name"]).mean()
G11_df = student_data.loc[(student_data["grade"]=="11th")].groupby(["school_name"]).mean()
G12_df = student_data.loc[(student_data["grade"]=="12th")].groupby(["school_name"]).mean()

# Combine the series into a dataframe
math_df = pd.DataFrame({"9th":G9_df["math_score"], "10th":G10_df["math_score"], "11th":G11_df["math_score"],
                        "12th":G12_df["math_score"]})
del math_df.index.name
math_df = math_df[["9th", "10th", "11th", "12th"]]
math_df
```

```python
# Optional: give the displayed data cleaner formatting
math_clean_df = pd.DataFrame({"9th":G9_df["math_score"].map('{:.1f}%'.format),
                              "10th":G10_df["math_score"].map('{:.1f}%'.format),
                              "11th":G11_df["math_score"].map('{:.1f}%'.format),
                              "12th":G12_df["math_score"].map('{:.1f}%'.format)
                             })
del math_clean_df.index.name
math_clean_df = math_clean_df[["9th", "10th", "11th", "12th"]]
math_clean_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

bin_df = school_performance_df
bin_df['bin'] = pd.cut(bin_df['Per Student Budget'].str.replace('$','').astype(float).round(), spending_bins, labels=group_names)
# bin_df['bin']
spend_bin_gb = bin_df.groupby(['bin'])

# Build results DF
scores_by_school_spending_df = pd.DataFrame({"Average Math Score":spend_bin_gb["Average Math Score"].mean(),
                                            "Average Reading Score":spend_bin_gb["Average Reading Score"].mean(),
                                            "% Passing Math":spend_bin_gb["% Passing Math"].mean(),
                                            "% Passing Reading":spend_bin_gb["% Passing Reading"].mean(),
                                            "% Overall Passing Rate":spend_bin_gb["% Overall Passing Rate"].mean(),
                                            })
scores_by_school_spending_df = scores_by_school_spending_df[["Average Math Score",
                                                             "Average Reading Score",
                                                             "% Passing Math",
                                                             "% Passing Reading",
                                                             "% Overall Passing Rate",
                                                            ]]
scores_by_school_spending_df.index.name = "Spending Ranges (Per Student)"
scores_by_school_spending_df
```

## Scores by School Size
